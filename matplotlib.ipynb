Q2: What is a scatter plot? Use the following code to generate data for x and y. Using this generated data
plot a scatter plot.

```python
import numpy as np
import matplotlib.pyplot as plt
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
```

A scatter plot is a type of plot that displays values for two variables for a set of data. The data is displayed as a collection of points, each having the value of one variable determining the position on the horizontal axis and the value of the other variable determining the position on the vertical axis¹. The scatter() method in the matplotlib library is used to draw a scatter plot¹.

```python
plt.scatter(x,y)
plt.xlabel('axis x')
plt.ylabel('axis y')
plt.title('Scatter Plot')
plt.show()
```

Q3: Why is the subplot() function used? Draw four line plots using the subplot() function.
Use the following data:

The subplot() function is used to create multiple plots in the same figure

```python
import numpy as np
x1 = np.array([0, 1, 2, 3, 4, 5]) 
y1 = np.array([0, 100, 200, 300, 400, 500])

x2 = np.array([0, 1, 2, 3, 4, 5]) 
y2 = np.array([50, 20, 40, 20, 60, 70])

x3 = np.array([0, 1, 2, 3, 4, 5]) 
y3 = np.array([10, 20, 30, 40, 50, 60])

x4 = np.array([0, 1, 2, 3, 4, 5]) 
y4 = np.array([200, 350, 250, 550, 450, 150])

#create a plot chart
plt.subplot(2,2,1)
plt.plot(x1,x2)

#create a pie chart
plt.subplot(2,2,2)
plt.plot(x2,y2)

#create a Bar chart
plt.subplot(2,2,3)
plt.plot(x3,y3)

#create a Histogram chart
plt.subplot(2,2,4)
plt.plot(x4,y4)
plt.show()
```

Q4: What is a bar plot? Why is it used? Using the following data plot a bar plot and a horizontal bar plot.

A bar plot is a common type of graphic used to show the relationship between a numeric and a categoric variable. It is used to represent categorical data, which is data that can be divided into distinct categories. In a bar plot, each category is represented by a separate bar, and the height of the bar represents the frequency or count of data points in that category.

```python
import numpy as np
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])
```

```python
#create a bar plot
plt.bar(company,profit,color = 'r',edgecolor = 'b')
```

```python
#create a horizontal bar plot.
plt.barh(profit,company)
```
